```python
import seaborn as sns
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc

palette = ['#ff3b30', '#ffcc00', '#4cd964', '#007aff', '#5856d6', '#ff9500', '#5ac8fa', '#ff2d55']
markers = ['o', '^', '<', 'd', '*', '2']
sns.set_style("ticks")
import matplotlib as mpl

mpl.rcParams['lines.markersize'] = 4
mpl.rcParams['lines.linewidth'] = 0.5
rc("font", **{'family': 'serif', 'serif': ['Arial'], 'size': 14})
plt.rcParams['font.family'] = 'Helvetica'
rc('text', usetex=True)
import os
import io
```

```python
basepath="./"
```

```python
def plot_time(total, name, title, use_logx=False, show_ylabel=True):
    basinhopping=[abs(float(line.split("\t")[0])) for line in total.split("\n") if len(line.split("\t")[0])>0]
    shgo=[abs(float(line.split("\t")[1])) for line in total.split("\n")if len(line.split("\t")[1])>0]
    dual_annealing=[abs(float(line.split("\t")[2])) for line in total.split("\n")if len(line.split("\t")[2])>0]
    Nelder_Mead=[abs(float(line.split("\t")[3])) for line in total.split("\n")if len(line.split("\t")[3])>0]
    BFGS=[abs(float(line.split("\t")[4])) for line in total.split("\n")if len(line.split("\t")[4])>0]
    CG=[abs(float(line.split("\t")[5])) for line in total.split("\n")if len(line.split("\t")[5])>0]
    
    
    df=pd.DataFrame({
        "model": basinhopping+shgo+dual_annealing+ Nelder_Mead+BFGS+CG,
        "type": ['Basinhopping']*len(basinhopping)+['SHGO']*len(shgo)+['Dual Annealing']*len(dual_annealing)+['Nelder Mead']*len(Nelder_Mead)+['BFGS']*len(BFGS)+['CG']*len(CG)
    })
    

    plt.figure(figsize=(6,4))
    ax=sns.boxplot(data=df, x="model", y="type", 
                   showfliers=False, 
                   width=.5,
                    palette=palette)
    ax.set_title(title, fontsize=16)
    if use_logx==True:
        plt.xscale('log', base=10)
    ax.set(xlabel="Time Usage (Mins)", ylabel="")
    if show_ylabel==False:
        ax.set(yticklabels=[])
        
  
    fname = os.path.join(basepath, name+"optimizers.time.pdf")
    plt.savefig(fname, bbox_inches='tight', pad_inches=0)
```
