```python
import seaborn as sns
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
palette=['#ff3b30', '#007aff', '#4cd964','#ffcc00', '#5856d6', '#ff9500', '#5ac8fa']
markers=['o', '^', '<', 'd', '*','2']
sns.set_style("ticks")
import matplotlib as mpl
mpl.rcParams['lines.markersize'] = 6
mpl.rcParams['lines.linewidth']=1
rc("font", **{'family':'serif','serif':['Palatino'], 'size':14})
rc('text', usetex=True)
import os
```

```python
basepath="./plots"
```

```python

def plot_gp_egp(total, name, title, metric_name, use_logy=False, show_ylabel=True):
    GP=[abs(float(line.split("\t")[0])) for line in total.split("\n")]
    EGP=[abs(float(line.split("\t")[1])) for line in total.split("\n")]
    VPG=[abs(float(line.split("\t")[2])) for line in total.split("\n")]
    PQT=[abs(float(line.split("\t")[3])) for line in total.split("\n")]
    DSR=[abs(float(line.split("\t")[4])) for line in total.split("\n")]
    GPMELD=[abs(float(line.split("\t")[5])) for line in total.split("\n")]
    Eureqa=[abs(float(line.split("\t")[6])) for line in total.split("\n")]
    df=pd.DataFrame({
        "model": EGP+GP+VPG+PQT+DSR+GPMELD+Eureqa,
        "type": ['CVGP']*10+['GP']*10+['VPG']*10+['PQT']*10+['DSR']*10+['GPMeld']*10+['Eureqa']*10
    })
    

    plt.figure(figsize=(3,3))
    ax=sns.boxplot(data=df, x="model", y="type", 
                   showfliers=True, 
                   width=0.5,
                    palette=palette)
    ax.set_title(title, fontsize=16)
    ax.set(xlabel=metric_name, ylabel="")
    if use_logy==True:
        plt.xscale('log', base=10)
        
    if show_ylabel==False:
        ax.set(yticklabels=[])
        

    fname = os.path.join(basepath, name+".pdf")
    plt.savefig(fname, bbox_inches='tight', pad_inches=0)
```
